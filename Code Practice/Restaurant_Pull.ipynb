```python
# Dependencies
# Dependencies
import pandas as pd
import numpy as np
import requests
import json

# Google API Key
from config import gkey
```

```python
types_df = pd.read_csv("../Resources/ethnic_restr.csv")
types_df.head()
```

```python
# set up additional columns to hold information
types_df['name'] = ""
types_df['address'] = ""
types_df['price_level'] = ""
types_df['rating'] = ""

types_df.head()
```

```python
# Find the closest restaurant to the coordinates

base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

# enter long&lat for desired city (this is San Francisco)
location = "37.78797,-122.43367"

params = {
    "location": location,  
    "rankby": "distance",
    "type": "restaurant",
    "key": gkey,
}
```

```python
# For loop to cycle through the restaruant ethnicity
for index, row in types_df.iterrows():
    # pull restaurant type and add to params 
    restr_type = row['ethnicity']
    params['keyword'] = restr_type
    
    # make API request
    print(f"Retrieving Results for Index {index}: {restr_type}.")
    response = requests.get(base_url, params=params).json()
    
    results = response['results']
    
    try:
        print(f"Closest {restr_type} restaurant is {results[0]['name']}.")
        
        types_df.loc[index, 'name'] = results[0]['name']
        types_df.loc[index, 'address'] = results[0]['vicinity']
        types_df.loc[index, 'price_level'] = results[0]['price_level']
        types_df.loc[index, 'rating'] = results[0]['rating']
        
    except (KeyError, IndexError):
        print("Missing field/result... skipping.")
        
    print("------------")
    
```
